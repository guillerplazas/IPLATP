````python
# GOOOOOOOOOD EXPERIMENT: MULTIPLE ISABELLE-CLIENTS
from isabelle_client import get_isabelle_client, start_isabelle_server
import asyncio
import nest_asyncio
import logging
from pprint import pprint

nest_asyncio.apply()


async def use_theories(client, theory_queue, master_dir):
    while not theory_queue.empty():
        theory = await theory_queue.get()
        try:
            print(f'Client {client.logger.name} processing {theory}')
            loop = asyncio.get_event_loop()
            response=await loop.run_in_executor(None, client.use_theories, [theory],None, master_dir)
            print(f'Client {client.logger.name} finished processing {theory}')
        finally:
            theory_queue.task_done()
        
            

async def shutdown_client(client):
    loop = asyncio.get_event_loop()
    await loop.run_in_executor(None, client.shutdown)

async def main():
    server_info, _ = start_isabelle_server()
    clients = [get_isabelle_client(server_info) for _ in range(4)]
    for i, client in enumerate(clients):
        client.logger = logging.getLogger(f'client_{i}')
        client.logger.setLevel(logging.INFO)
        client.logger.addHandler(logging.FileHandler(f'session_{i}.log'))
    theory_files = [f"aux_{i}" for i in range(9)]
    theory_queue = asyncio.Queue()
    for theory in theory_files:
        await theory_queue.put(theory)

    tasks = [asyncio.create_task(use_theories(client, theory_queue, './auxs')) for client in clients]
    
    await asyncio.gather(*tasks)
    
    
    await theory_queue.join()  
    await asyncio.gather(*[shutdown_client(client) for client in clients])

asyncio.run(main())
````
